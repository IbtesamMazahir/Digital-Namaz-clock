```python
import tkinter as tk
from tkinter import *
import time

background_color = input("Enter Background color: ")
text_color = input("Enter Text color: ")

root=tk.Tk()
root.geometry("800x800")
root.title("Namaz Clock")
root.configure(bg="black")

def tick(time1=""):
    time2= time.strftime("%I : %M : %S")
    clock_frame.configure(text=time2)
    clock_frame.after(20,tick)

def Fajar_namaz():
    Fajar.configure(text="04:45 AM Fajar")
    
def Zuhar_namaz():
    Zuhar.configure(text="01:00 PM Zuhar")
    
def Asar_namaz():
    Asar.configure(text="05:30 PM Asar")
    
def Maghrib_namaz():
    Maghrib.configure(text="07:35 PM Maghrib")
    
def Isha_namaz():
    Isha.configure(text="08:27 PM Isha")
    
def Jummah_namaz():
    Jummah.configure(text="02:00 PM Jummah")
    
    
clock_frame = Label(root, font=("Arial",50,"bold"),bg=background_color, fg=text_color)
clock_frame.pack(fill=X)

Fajar = Label(root, font=("Arial",25,"bold"),bg=background_color, fg=text_color,anchor='n')
Fajar.pack(fill=X)


Zuhar = Label(root, font=("Arial",25,"bold"),bg=background_color, fg=text_color,anchor='n')
Zuhar.pack(fill=X)


Asar = Label(root, font=("Arial",25,"bold"),bg=background_color, fg=text_color,anchor='n')
Asar.pack(fill=X)


Maghrib = Label(root, font=("Arial",25,"bold"),bg=background_color, fg=text_color,anchor='n')
Maghrib.pack(fill=X)

Isha = Label(root, font=("Arial",25,"bold"),bg=background_color, fg=text_color,anchor='n')
Isha.pack(fill=X)

Jummah = Label(root, font=("Arial",25,"bold"),bg=background_color, fg=text_color,anchor='n')
Jummah.pack(fill=X)

tick()
Fajar_namaz()
Zuhar_namaz()
Asar_namaz()
Maghrib_namaz()
Isha_namaz()
Jummah_namaz()

root.mainloop()
```
